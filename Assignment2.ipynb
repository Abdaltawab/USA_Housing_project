```python
#cost function in regression consider the Accuracy
```

```python
def computCost(x,y,theta):
    z=np.power(((x*theta.T)-y),2)
    return np.sum(z)/(2*len(x))
```

```python
print('Accuracy = ',computCost(x,y,theta))
```

```python
#calc Grident Decent to calc the final weights
```

```python
def gradientDesent(x,y,theta,alpha,iters):
    temp=np.matrix(np.zeros(theta.shape))
    parameters=int(theta.ravel().shape[1])
    cost=np.zeros(iters)
    
    for i in range(iters):
        error= (x*theta.T)-y
        for j in range(parameters):
            term=np.multiply(error,x[:,j])
            term[0,j]=theta[0,j]-((alpha/len(x))*np.sum(term))
            theta=temp
            cost[i]=computCost(x,y,theta)
            return theta,cost
        

    
```

```python
alpha=0.01
iters=190000
```

```python

g,cost=gradientDesent(x,y,theta,alpha,iters)
```

```python
print('the final weights = \n',g)
```
